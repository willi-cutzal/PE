### **problema 1** ###

```python
number = 1000
multiples = []
for i in range(1, number):
    if (i % 3 == 0) or (i % 5 == 0):
        multiples.append(i)
    
#print(multiples)
print(f"la suma de los múltiplos de 3 y de 5 es: {sum(multiples)}")
```
